Next step is to convert the 'Start Time' to pandas datetime format and then convert to Canada/Eastern timezone. I use dft.dtypes to check if my 'Start Time' was succesfully converted.

I first make the 'Start Time' the index as the pandas tz.convert() function only works on index. It turned out to be a task that was temperamental as it kept giving me attribute error for tz_convert() function. 

I then reset the index back to the default index.

Then I convert 'Duration' to pandas time duration format: timedelta.

```python
dft['Start Time'] = pd.to_datetime(dft['Start Time'], utc = True)
dft.dtypes
```

```python
dft = dft.set_index('Start Time')
dft.head()
```

```python
dft.index = dft.index.tz_convert('Canada/Eastern')
```

```python
dft.head(5)
```

```python
dft = dft.reset_index()
```

```python
dft.head(5)
```

```python
dft['Duration'] = pd.to_timedelta(dft['Duration'])
dft.dtypes
```

I then convert 'Duration' into seconds because this will make my statistical analysis simpler.

So I use the timedelta function .total_seconds() to get a new column that shows the total seconds I watch each episode.

```python
dft['DurationSeconds'] = dft['Duration'].dt.total_seconds()
```

```python
dft.head(5)
```

I wish to see how my netflix viewing activity changes from each day of the week. So I will extract day, month, year from the 'Start Time' column using datetime accessors.

It is important to know here that pandas datetime series accessor for weekday shows integer values where: The day of the week is represented as Monday=0, Sunday=6. 
mon = 0
tues = 1
wed = 2
thurs = 3
friday = 4
sat = 5
sun = 6

```python
dft['Day'] = dft['Start Time'].dt.weekday
dft['Year'] = dft['Start Time'].dt.year

dft.loc[dft['Day'] == 0]
```

```python
avg_time_per_day = dft.groupby('Day').DurationSeconds.mean().reset_index()
```

```python
#avg_time_per_day.sort_index()

day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_time_per_day['Days'] = day_of_week
avg_time_per_day
```

```python

sns.set_style('darkgrid')
font = {'family': 'serif',
        'color': '#004466',
        'weight': 'normal',
        'size': 16}
plt.figure(figsize=(20,15))
grph = sns.barplot(x='Days', y='DurationSeconds', data=avg_time_per_day.reset_index())
grph.set_ylabel('Viewing Duration (in Seconds)',fontdict={'size': 13, 'family': 'serif'})
grph.set_xlabel('Day of Week', fontdict={'size': 13, 'family': 'serif'})
grph.set_title('Average Viewing Time Per Day(seconds)', fontdict = font)
grph.tick_params(axis = 'both', labelsize= 12)

```

Next step is to see what my top 10 shows have been. I will first have to split the Title string

I was having trouble splitting the string in the Title column because the strings in each row have varied numbers of ':' delimiter that I wanted to split across. So i limited the split to n = 1, so it only splits once. This way I get the Show name. For movies I get NaN values in Season column

```python
df_title = dft
df_title.head(5)
df_title[['Show', 'Season']] = df_title.Title.str.split(':', n = 1, expand = True)
   
df_count = df_title.groupby('Show').size().reset_index(name = 'count')
df_count = df_count.sort_values('count', ascending = False)
df_count.head(5)
```

```python
df_seconds = df_title.groupby('Show')['DurationSeconds'].sum().reset_index().sort_values('DurationSeconds', ascending =False)
df_seconds.head(10)
```

```python
sns.set_style('darkgrid')
font = {'family': 'serif',
        'color': '#004467',
        'weight': 'normal',
        'size': 16}
plt.figure(figsize=(20,15))
grph = sns.barplot(x= df_seconds['Show'][:10], y=df_seconds['DurationSeconds'][:10], data=df_seconds)
grph.set_ylabel('Total Viewing Time(seconds)',fontdict={'size': 13, 'family': 'serif'})
grph.set_xlabel('Show', fontdict={'size': 13, 'family': 'serif'})
grph.set_title('My top shows', fontdict = font)
grph.tick_params(axis = 'both', labelsize= 13)
```
